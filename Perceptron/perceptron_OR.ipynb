```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np

# Proper 2D array definition
x = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y=np.array([0,1,1,1])
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10
print("intial weight:",w)
print("intial bias",b)
```

```python
def step_function(z):
    return 1 if z>=0 else 0
    
```

```python
for epoch in range(epochs):
    print("Epoch", epoch)
    
    for j in range(len(x)):
        # 1. Compute weighted sum
        z = np.dot(w, x[j]) + b
        
        # 2. Apply activation function
        y_pred = step_function(z)
        
        # 3. Compute error
        error = y[j] - y_pred
        
        # 4. Update weights
        w = w + learning_rate * error * x[j]
        
        # 5. Update bias
        b = b + learning_rate * error
        
        print(f"Input:{x[j]}, Predicted:{y_pred}, Actual:{y[j]}, w:{w}, b:{b}")
```

```python
print("final weight:",w)
print("final bias:",b)
```

```python
print("\n testing or gate")
for i in range(len(x)):
    z=np.dot(w,x[i])+b
    prediction=step_function(z)
    print("Input:",x[i],"prediction:",prediction)
```

```python
# Plot points
for i in range(len(x)):
    if y[i] == 0:
        plt.scatter(x[i][0], x[i][1], marker='o')
    else:
        plt.scatter(x[i][0], x[i][1], marker='x')

# Plot decision boundary
x1_vals = np.linspace(-0.5, 1.5, 100)

if w[1] != 0:
    x2_vals = -(w[0]*x1_vals + b) / w[1]
    plt.plot(x1_vals, x2_vals)

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron Decision Boundary (or Gate)")
plt.grid(True)
plt.show()
```
