```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np

# Dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y=np.array([0,1,1,0])
```

```python
#neuron 1
w1=1
w2=1
b1=-0.5
#neuron 2
w3=1
w4=1
b2=-1.5
#neuron 3
w5=1
w6=-2
b3=-0.5
```

```python


# Step function
def StepFunction(x):
    return 1 if x >= 0 else 0
```

```python
def forward_propagation(x1, x2):

    # Hidden layer
    h1 = w1*x1 + w2*x2 + b1
    a1 = StepFunction(h1)

    h2 = w3*x1 + w4*x2 + b2
    a2 = StepFunction(h2)

    # Output layer
    y = w5*a1 + w6*a2 + b3
    output = StepFunction(y)

    return output


print("XOR Results using ANN Forward Propagation\n")
```

Test Forward Propagation for All XOR Inputs

```python
print("XOR Results using ANN Forward Propagation\n")

for x in X:
    prediction = forward_propagation(x[0], x[1])
    print(f"Input: {x}  Output: {prediction}")
```

```python

for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], marker='o')
    else:
        plt.scatter(X[i][0], X[i][1], marker='x')

# Create x values
x_vals = np.linspace(-0.5, 1.5, 100)

# OR boundary line
x2_or = -(w1*x_vals + b1) / w2
plt.plot(x_vals, x2_or, linestyle='--', label="OR Boundary")

# AND boundary line
x2_and = -(w3*x_vals + b2) / w4
plt.plot(x_vals, x2_and, linestyle='-', label="AND Boundary")

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("XOR Decision Boundaries (Hidden Layer)")
plt.legend()
plt.grid(True)
plt.show()
```
